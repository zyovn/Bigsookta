```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['target'] = iris.target
```

```python
iris_df['target'] = iris_df['target'].map({0:"setisa", 1:"versicolor", 2:"virgirinica"})
```

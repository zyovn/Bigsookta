```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
iris = load_iris()
iris_df = pd.DataFrame(data = iris.data , columns = iris.feature_names)
iris_df["target"] = iris.target
iris_df["target"] = iris_df["target"].map({0:"setosa" , 1:"versicolor" , 2:"virginica"})
```

```python
plt.plot([1, 2, 3],[30, 40, 30],'k')
plt.plot([4, 5, 6],[20, 10, 20],'r')
plt.plot([7, 8, 9],[30, 40, 30],'k')
plt.axis([0.5, 9.5, 9, 41])
plt.show()
```

```python
x = iris_df["target"]
y = iris_df["petal length (cm)"]
plt.bar(x,y)
```

```python
data = [iris_df[iris_df["target"] == "setosa"]["petal length (cm)"],
        iris_df[iris_df["target"] == "versicolor"]["petal length (cm)"],
        iris_df[iris_df["target"] == "virginica"]["petal length (cm)"]]
               
plt.boxplot(data, labels=["setosa", "versicolor", "virginica"], showmeans=True)
plt.show()
```

```python
x = ['많이 감소', '조금 감소','변화없음','조금 증가', '많이 증가','해당없음']
y = [20, 61,449, 179, 67,224]
line_graph = plt.plot(x, y, 'o-', color = 'red')
plt.ylim(0, 500)

for i in range(len(x)):
    height = y[i]
    plt.text(x[i], height + 0.25, '%.1f' %height, ha='center', va='bottom', size = 12)

plt.title('코로나19로 인한 학습변화')

plt.show()
```
